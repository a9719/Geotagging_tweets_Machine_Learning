```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.metrics import classification_report
```

```python
def preprocess(train,dev):
    train=pd.read_csv(train,index_col=0)
    dev=pd.read_csv(dev,index_col=0)
    x_train=train[train.columns[:-1]]
    y_train=train['Location']
    
    x_dev=dev[dev.columns[:-1]] 
    y_dev=dev['Location']
    
    return x_train,y_train,x_dev,y_dev
    
```

```python
x_train,y_train,x_dev,y_dev=preprocess("train-top10.csv","dev-top10.csv")
```

```python
#WORKING ON TOP10 FEATURED FILES
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
NV_model=model.fit(x_train,y_train)
pred = NV_model.predict(x_dev)

accuracy = accuracy_score(y_dev, pred)
print('Accuracy',100*accuracy,'%')

labels = ['Sydney', 'Melbourne', 'Brisbane', 'Perth']
matrix=confusion_matrix(y_dev, pred,labels)
cm = pd.DataFrame(matrix, index=labels, columns=labels)
sns.heatmap(cm, annot=True, fmt='d',cmap="BuPu")
plt.title("Confusion Matrix")
plt.xlabel("Predicted classes")
plt.ylabel("Actual classes")
plt.yticks(rotation=0)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
GNB_model=model.fit(x_train,y_train)
pred = GNB_model.predict(x_dev)

accuracy = accuracy_score(y_dev, pred)
print('Accuracy',100*accuracy,'%')

labels = ['Sydney', 'Melbourne', 'Brisbane', 'Perth']
matrix=confusion_matrix(y_dev, pred,labels)
cm = pd.DataFrame(matrix, index=labels, columns=labels)
sns.heatmap(cm, annot=True, fmt='d',cmap="BuPu")
plt.title("Confusion Matrix")
plt.xlabel("Predicted classes")
plt.ylabel("Actual classes")
plt.yticks(rotation=0)
plt.show()
```

```python
import nltk
from string import punctuation
import re
from nltk.tokenize import TweetTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#train data raw
train = pd.read_csv('train-raw.tsv', sep='\\t').iloc[1:]
train.columns = ['InstanceID','location', 'tweet']
x_train=train['tweet']
y_train=train['location']


#dev data raw
dev = pd.read_csv('dev-raw.tsv',sep='\\t').reset_index()
dev.drop('level_0', axis=1 ,inplace=True)
dev.columns = ['location' ,'tweet']
x_dev=dev['tweet']
y_dev=dev['location']


#test data raw
test = pd.read_csv('test-raw.tsv', sep='\\t').reset_index()
test.drop('level_0', axis=1, inplace=True)
test.columns = ['location', 'tweet']
x_test=test['tweet']
y_test=test['location']
```

```python
lmtzr = nltk.WordNetLemmatizer()
tknzr = TweetTokenizer(strip_handles=True, reduce_len=True)
stopwords = set(stopwords.words('english'))
def feature(text):
    text = ''.join([i.casefold() for i in text if i not in punctuation])
    tokens = [lmtzr.lemmatize(word) for word in text.split() if word not in stopwords]
    text= ' '.join(tokens)
    text= ''.join([w if ord(w) < 128 else ' ' for w in text])
    text = text.lower()
    text = re.sub(r'^https?:\/\/.*[\r\n]*', '', text)
    
    return tknzr.tokenize(text)
```

```python
#https://adataanalyst.com/scikit-learn/countvectorizer-sklearn-example/   vectorizer method adapted from
vectorizer = CountVectorizer(analyzer=feature).fit(x_train)
```

```python
#vectorizing
x_train=vectorizer.transform(x_train)
x_dev=vectorizer.transform(x_dev)
x_test = vectorizer.transform(x_test)
```

```python
#MULTINOMIALNB
model = MultinomialNB()
NV_model=model.fit( (x_train),y_train)
y_pred = NV_model.predict((x_dev))
accuracy = accuracy_score(y_dev,y_pred)

print('Accuracy',100*accuracy,'%')

labels = ['Sydney', 'Melbourne', 'Brisbane', 'Perth']
matrix=confusion_matrix(y_dev, y_pred,labels)
cm = pd.DataFrame(matrix, index=labels, columns=labels)
sns.heatmap(cm, annot=True, fmt='d',cmap="BuPu")
plt.title("Confusion Matrix")
plt.xlabel("Predicted classes")
plt.ylabel("Actual classes")
plt.show()
```

```python
model= SGDClassifier(max_iter=5)
SGD_model=model.fit( (x_train),y_train)
y_pred = SGD_model.predict((x_dev))
accuracy = accuracy_score(y_dev,y_pred)

print('Accuracy',100*accuracy,'%')

labels = ['Sydney', 'Melbourne', 'Brisbane', 'Perth']
matrix=confusion_matrix(y_dev, y_pred,labels)
cm = pd.DataFrame(matrix, index=labels, columns=labels)
sns.heatmap(cm, annot=True, fmt='d',cmap="BuPu")
plt.title("Confusion Matrix")
plt.xlabel("Predicted classes")
plt.ylabel("Actual classes")
plt.show()
```
